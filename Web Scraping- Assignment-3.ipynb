```python
# Defining a function
def phone_info(item):
    model=item.find('div',{'class':"_4rR01T"}).text[0:item.find('div',{'class':"_4rR01T"}).text.find('(')]
    star=item.find('div',{'class':"_3LWZlK"}).text
    ram=item.find('li',{'class':"rgWa7D"}).text[0:item.find('li',{'class':"rgWa7D"}).text.find('|')]
    rom=item.find('li',{'class':"rgWa7D"}).text[item.find('li',{'class':"rgWa7D"}).text.find('|')+1:][0:10].strip()
    display=item.find_all('li')[1].text.strip()
    camera=item.find_all('li')[2].text.strip()
    battery=item.find_all('li')[3].text.strip()
    processor=item.find_all('li')[4].text.strip()
    price=item.find('div',{'class':'_30jeq3 _1_WHN1'}).text
    result=(model,star,ram,rom,display,camera,battery,processor,price)
    return result
```

```python
record_list=[]
results=soup.find_all('a',{'class':"_1fQZEK"})
for item in results:
    record_list.append(phone_info(item))
```

```python
pd.DataFrame(record_list,columns=['model','star','ram','rom','display','camera','battery','processor','price'])
```

```python
Name_tag=driver.find_elements_by_xpath("//a[@class='title title-6']")
for i in Name_tag:
    name=i.text
    Hostel_name.append(name)
    
dist_tag=driver.find_elements_by_xpath("//span[@class='description']")
for i in dist_tag:
    distance=i.text
    DistFromCity.append(distance)
    
rating_tag=driver.find_elements_by_xpath("//div[@class='score orange big']")
for i in rating_tag:
    rating=i.text
    Rating.append(rating)
    
TotalReview_tag=driver.find_elements_by_xpath("//div[@class='reviews']")
for i in TotalReview_tag:
    review=i.text
    Total_Review.append(review)
    
overall_Reviews_tag=driver.find_elements_by_xpath("//div[@class='keyword']")
for i in overall_Reviews_tag:
    overall=i.text
    Overall_reviews.append(overall)
    
private_tag=driver.find_elements_by_xpath("//p[@class='no-price body-3 no-privates']")
for i in private_tag:
    private=i.text
    Private.append(private)

dorms_tag=driver.find_elements_by_xpath("//div[@class='price title-5']")
for i in dorms_tag:
    dorms=i.text
    Dorms_price.append(dorms)
    
facility_tag=driver.find_elements_by_xpath("//span[@class='label']")
for i in facility_tag:
    facility=i.text
    Facitily.append(facility)

property_tag=driver.find_elements_by_xpath("//span[@class='rating body-3']")
for i in property_tag:
    prop=i.text
    Property_Desc.append(prop)
```

```python
hostel_detail=pd.DataFrame({})
hostel_detail['Hostel_name']=Hostel_name[:8]
hostel_detail['DistFromCity']=DistFromCity[:8]
hostel_detail['Rating']=Rating[:8]
hostel_detail['Total_Review']=Total_Review[:8]
hostel_detail['Overall_reviews']=Overall_reviews[:8]
hostel_detail['Private']=Private[:8]
hostel_detail['Dorms_price']=Dorms_price[:8]
hostel_detail['Facitily']=Facitily[:8]
hostel_detail['Property_Desc']=Property_Desc[:8]


hostel_detail
```
