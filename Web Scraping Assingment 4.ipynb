### 2.Scrape the details team India’s international fixtures from bcci.tv.
Url = https://www.bcci.tv/.
You need to find following details:

A) Match title (I.e. 1st ODI)
B) Series
C) Place
D) Date
E) Tim

```python
# import all necessary libraries

import selenium
import pandas as pd
from selenium import webdriver
```

```python
# Open the driver

driver=webdriver.Chrome(r"C:/Users/LENOVO/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
# Use the required URL with the driver
driver.get('https://www.bcci.tv/')
```

```python
## Need to extract all details
Match_title=[]
Series=[]
Place=[]
Date=[]
Time=[]

```

```python
Match_tag=driver.find_elements_by_xpath("//span[@class='ng-binding']")
for i in Match_tag:
    match=i.text
    Match_title.append(match)

Series_tag=driver.find_elements_by_xpath("//p[@class='match-place ng-binding']")
for i in Series_tag:
    series=i.text
    Series.append(series)
    
place_tag=driver.find_elements_by_xpath("//span[@class='ng-binding ng-scope']")
for i in place_tag:
    place=i.text
    Place.append(place)
    
date_tag=driver.find_elements_by_xpath("//div[@class='match-card-left match-schedule']")
for i in date_tag:
    date=i.text
    Date.append(date)
    
Time_tag=driver.find_elements_by_xpath("//div[@class='match-card-right match-schedule ']")
for i in Time_tag:
    time=i.text
    Time.append(time)
```

```python
BCCI=pd.DataFrame({})
BCCI['Match_title']=Match_title[:14]
BCCI['Series']=Series[:14]
BCCI['Place']=Place[:14]
BCCI['Date']=Date[:14]
BCCI['Time']=Time[:14]
BCCI
```

```python
driver.close()
```

# 5 Scrape the details of trending repositories on Github.com.
Url = https://github.com/

You have to find the following details: A) Repository title B) Repository description C) Contributors count D) Language used

```python
# import all necessary libraries

import selenium
import pandas as pd
from selenium import webdriver
```

```python
# Open the driver

driver=webdriver.Chrome(r"C:/Users/LENOVO/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
driver.get('https://github.com/trending')
```

```python
# A) Repository title B) Repository description C) Contributors count D) Language used

Title = []
RepositoryDesc = []
Contributor = []
Lang = []
```

```python
# Scrap Title
try:
    title_tag=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']")
    for i in title_tag:
        title=i.text
        Title.append(title)
except:
    Title.append('-')
    
# Scrap RepositoryDesc
try:
    rep_tag=driver.find_elements_by_xpath("//p[@class='col-9 color-fg-muted my-1 pr-4']")
    for i in rep_tag:
        rep=i.text
        RepositoryDesc.append(rep)
except:
    RepositoryDesc.append('-')

# Scrap Contributor
try:
    Con_tag=driver.find_elements_by_xpath("//a[@class='Link--muted d-inline-block mr-3']")
    for i in Con_tag:
        con=i.text
        Contributor.append(con)
except:
    Contributor.append('-')
    

# Scrap Lang
try:
    lang_tag=driver.find_elements_by_xpath("//span[@itemprop='programmingLanguage']")
    for i in lang_tag:
        lang=i.text
        Lang.append(lang)
except:
    Lang.append('-')
```

```python
Repository=pd.DataFrame({})

Repository['Title']=Title[:23]
Repository['RepositoryDesc']=RepositoryDesc[:23]
Repository['Contributor']=Contributor[:23]
Repository['Lang']=Lang[:23]
Repository
```

```python
driver.close()
```

# 9 Scrape the details most watched tv series of all time from imdb.com.

Url = https://www.imdb.com/list/ls095964455/

You have to find the following details: A) Name B) Year span C) Genre D) Run time E) Ratings F) Votes

```python
# import all necessary libraries

import selenium
import pandas as pd
from selenium import webdriver
import time
```

```python
# Open the driver

driver=webdriver.Chrome(r"C:/Users/LENOVO/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
# Use the required URL with the driver
driver.get('https://www.imdb.com/list/ls095964455/')
```

```python
## Need to extract all details
Name=[]
Yearspan=[]
Genre=[]
Runtime=[]
Ratings=[]
Votes=[]
```

```python
# Scrap Name
try:
    name_tag=driver.find_elements_by_xpath("//h3[@class='lister-item-header']")
    for i in name_tag:
        name=i.text
        Name.append(name)
except:
    Name.append('-')
    
# Scrap Yearspan
try:
    year_tag=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
    for i in year_tag:
        year=i.text
        Yearspan.append(year)
except:
    Yearspan.append('-')
    
# scrap Genre
try:
    genre_tag=driver.find_elements_by_xpath("//span[@class='genre']")
    for i in genre_tag:
        genre=i.text
        Genre.append(genre)
except:
    Genre.append('-')

# Runtime 
try:
    runtime_tag=driver.find_elements_by_xpath("//span[@class='runtime']")
    for i in runtime_tag:
        runtime=i.text
        Runtime.append(runtime)
except:
    Runtime.append('-')
    
# Ratings 
try:
    ratings_tag=driver.find_elements_by_xpath("//div[@class='ipl-rating-widget']")
    for i in ratings_tag:
        ratings=i.text
        Ratings.append(ratings)
except:
    Ratings.append('-')
    
# Votes 
try:
    vote_tag=driver.find_elements_by_xpath("//span[@name='nv']")
    for i in vote_tag:
        vote=i.text
        Votes.append(vote)
except:
    Votes.append('-')
```

```python
Shows=pd.DataFrame({})
Shows['Name']=Name[:100]
Shows['Yearspan']=Yearspan[:100]
Shows['Genre']=Genre[:100]
Shows['Runtime']=Runtime[:100]
Shows['Ratings']=Ratings[:100]
Shows['Votes']=Votes[:100]
Shows
```
